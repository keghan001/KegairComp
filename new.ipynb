```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
#Importing KegAir Data
data_dir = os.getcwd()
os.listdir(data_dir)

data = pd.read_csv('data/kegAir.csv', index_col=None, parse_dates=[0], dayfirst=True)

data['DateTime'] = pd.to_datetime(data['DateTime'], errors='coerce', dayfirst=True)
data.dropna(subset=['DateTime'], inplace=True)

data.set_index('DateTime', inplace=True)
```

```python
data.shape
```

```python
start_dt = pd.to_datetime('2024-05-29 15:30:00')
end_dt = pd.to_datetime('2024-06-05 10:12:00')

# data.loc['2024-05-29 15:29:00']
#Time range mask
mask = (data.index >= start_dt) & (data.index <= end_dt)
new_data = data.loc[mask]
new_data.shape
```

```python
new_data.info()
```

```python
data.info()
```

```python
# resampling data to avearage it every 2min
resampled_data = data.resample('2T').mean()

new_resample = new_data.resample('2T').mean()
```

```python
new_resample.dropna(axis= 0, inplace=True)
new_resample.info()
new_resample.tail(10)
```

```python
#Cleaning out extrapolated data with original datetime
start_datetime = pd.to_datetime('2024-05-29 15:29:00')
end_datetime = pd.to_datetime('2024-06-05 10:14:00')

# Step 1: Filter the original DataFrame to only include data within the datetime range
filtered_data = data[(data.index >= start_datetime) & (data.index <= end_datetime)]

resampled_data = filtered_data.resample('2T').mean()

#
resampled_data = resampled_data[(resampled_data.index >= start_datetime) & (resampled_data.index <= end_datetime)]
```
